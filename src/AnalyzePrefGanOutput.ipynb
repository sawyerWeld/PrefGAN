```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
torch_data = torch.load('test_votes_big')
```

```python
data = []
for sub_li in torch_data:
    py_li = [a.tolist() for a in sub_li]
    data.append(py_li)
```

```python
def dif(a,b):
    delta = 0
    for i in range(len(a)):
        delta += abs(a[i]-b[i])
    return delta
```

```python
graph_difs = []
for graph in data:
    difs = []
    for i in range(len(graph)-1):
        difs.append(dif(graph[i],graph[i+1]))
    graph_difs.append(difs)
```

```python
%matplotlib inline
import mpld3
mpld3.enable_notebook()
```

```python
palette = plt.get_cmap('Set1')
plt.title('plots')
for i, graph in enumerate(graph_difs):
    plt.plot(list(range(len(graph))),graph,label=i,color=palette(i),alpha=0.5)
plt.show()
```
