```python
from graphviz import Digraph
```

```python
g = Digraph('results graph 1')

g.node('Alt 1')
g.node('Alt 2')
g.node('Alt 3')
g.node('Alt 4')
g.node('Alt 5')
g.node('Alt 6')

g.edge('Alt 1', 'Alt 2')
for i in range(3,7):
    g.edge('Alt 1', 'Alt %d' % i)
    g.edge('Alt 2', 'Alt %d' % i)
g
```
